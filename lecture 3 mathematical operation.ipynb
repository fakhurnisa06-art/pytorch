```python
import torch
import numpy as np
```

**Tensor Math operation**

```python
tensor_a=torch.tensor([2,4,6,8])
tensor_b=torch.tensor([1,3,5,7])
```

```python
#addition
tensor_a+tensor_b
```

```python
#addition long hand
torch.add(tensor_a,tensor_b)
```

```python
#sub function
torch.sub(tensor_a,tensor_b)
```

```python
#Multiplication
tensor_a*tensor_b
```

```python
#Multiplication long hand
torch.mul(tensor_a,tensor_b)
```

```python
#Divison
tensor_a/tensor_b
```

```python
#long hand divion
torch.div(tensor_a,tensor_b)
```

```python
#Long reminder module
torch.remainder(tensor_a,tensor_b)
```

```python
#reminder module
tensor_a % tensor_b
```

```python
# Exponent / power
torch.pow(tensor_a,tensor_b)
```
